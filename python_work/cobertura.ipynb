Verificando a diferença. Operador de diferença (!=).

```python
requerendo_cobertura = 'cogumelo'

if requerendo_cobertura != 'anchovas':
    print('Coloque anchovas por favor!')
```

```python
requerendo_cobertura = 'anchovas'

if requerendo_cobertura != 'cogumelo':
    print('Coloque cogumelo por favor!')
```
