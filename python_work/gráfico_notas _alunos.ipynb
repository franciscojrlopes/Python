```python
import matplotlib.pyplot as plt


notas = [ 1, 5, 6, 3, 3, 7 ]
grupos = ['Grupo 01', 'Grupo 02', 'Grupo 03', 'Grupo 04', 'Grupo 05', 'Grupo 06']
plt.xlabel('Grupos')
plt.ylabel('Notas')
plt.title('Notas por grupo raspadinha de 23/02/2024')

plt.bar(grupos, notas)
```
