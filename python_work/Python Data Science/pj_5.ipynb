Groupby agrupamdo dados no df:

```python
import pandas as pd
import re
df = pd.read_csv('house.csv')
df
```

```python
df.groupby(['SellerG']).mean()
```

```python
df.groupby(['SellerG']).sum().sort_values('Price', ascending= False)
```

```python
df.groupby(['SellerG']).sum().sort_values('Price', ascending= False).head(10)
# Dez maiores vendedores............
```

```python
df.groupby(['SellerG']).count().sort_values('Price', ascending= False).head(10)

# Count por preço
```

```python
df.groupby(['SellerG']).count().sort_values('Rooms', ascending= False).head(10)

# Count por quartos.
```

```python
df.groupby(['SellerG']).count()
```
