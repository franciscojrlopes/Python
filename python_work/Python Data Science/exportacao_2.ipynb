```python
import pandas as pd
tabela = pd.read_csv('exportacoes_franca.csv', encoding=False)
display(tabela)
```

```python
tabela_de_produtos_exportados = tabela[['SH2 Description', 'US$ FOB']].groupby('SH2 Description').sum()
display(tabela_de_produtos_exportados)
```

```python
tabela_de_produtos_exportados = tabela[['SH2 Description', 'US$ FOB']].groupby('SH2 Description').sum()
tabela_de_produtos_exportados = tabela_de_produtos_exportados.sort_values(by='US$ FOB',ascending=False)
display(tabela_de_produtos_exportados)
```
